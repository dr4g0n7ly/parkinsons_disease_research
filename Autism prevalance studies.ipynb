```python
df1 = pd.read_csv('Datasets/Eye-Tracking Dataset/Eye-tracking Output/1.csv')
dfm = pd.read_csv('Datasets/Eye-Tracking Dataset/Metadata_Participants.csv')
```

```python
def toNumeric(x):
    try:
        return int(x)
    except:
        if (x == 'Unidentified(Neg)'):
            return -1
        elif (x == 'Unidentified(Pos)'):
            return -2
        elif (x == '-'):
            return 0
        
df1['Participant'] = df1['Participant'].apply(lambda x: toNumeric(x) + 1)

df1['Index_Right'] = df1['Index Right'].apply(lambda x: toNumeric(x))
df1['Index_Left'] = df1['Index Left'].apply(lambda x: toNumeric(x))

df1['Pupil_Diameter_Right_[mm]'] = df1['Pupil Diameter Right [mm]'].apply(lambda x: toNumeric(x))
df1['Pupil_Diameter_Left_[mm]'] = df1['Pupil Diameter Left [mm]'].apply(lambda x: toNumeric(x))

df1['Point_of_Regard_Right_X_[px]'] = df1['Point of Regard Right X [px]'].apply(lambda x: toNumeric(x))
df1['Point_of_Regard_Right_Y_[px]'] = df1['Point of Regard Right Y [px]'].apply(lambda x: toNumeric(x))
df1['Point_of_Regard_Left_X_[px]'] = df1['Point of Regard Left X [px]'].apply(lambda x: toNumeric(x))
df1['Point_of_Regard_Left_Y_[px]'] = df1['Point of Regard Left Y [px]'].apply(lambda x: toNumeric(x))

df1['Gaze_Vector_Left_X'] = df1['Gaze Vector Left X'].apply(lambda x: toNumeric(x))
df1['Gaze_Vector_Left_Y'] = df1['Gaze Vector Left Y'].apply(lambda x: toNumeric(x))
df1['Gaze_Vector_Left_Z'] = df1['Gaze Vector Left Z'].apply(lambda x: toNumeric(x))
df1['Gaze_Vector_Right_X'] = df1['Gaze Vector Right X'].apply(lambda x: toNumeric(x))
df1['Gaze_Vector_Right_Y'] = df1['Gaze Vector Right Y'].apply(lambda x: toNumeric(x))
df1['Gaze_Vector_Right_Z'] = df1['Gaze Vector Right Z'].apply(lambda x: toNumeric(x))

df1['Mouse_Position_X_[px]'] = df1['Mouse Position X [px]'].apply(lambda x: toNumeric(x))
df1['Mouse_Position_Y_[px]'] = df1['Mouse Position Y [px]'].apply(lambda x: toNumeric(x))

df1['Scroll_Direction_X'] = df1['Scroll Direction X'].apply(lambda x: toNumeric(x))
df1['Scroll_Direction_Y'] = df1['Scroll Direction Y'].apply(lambda x: toNumeric(x))
```

```python
df1 = pd.merge(df1, dfm, on='Participant', how='left')
df1.loc[df1.Participant == -1, 'Class'] = "TD"
df1.loc[df1.Participant == -2, 'Class'] = "ASD"
df1.sort_values(by=['Participant']).head()
```

```python
def proc_data(df):
    
    df['Category_Group'] = df['Category Group']
    df['Category_Group'] = pd.Categorical(df.Category_Group)
    
    df['Category_Left'] = df['Category Left']
    df['Category_Left'] = pd.Categorical(df.Category_Left)
    
    df['Category_Right'] = df['Category Right']
    df['Category_Right'] = pd.Categorical(df.Category_Right)
    
    df['AOI_Name_Left'] = df['AOI Name Left']
    df['AOI_Name_Left'] = pd.Categorical(df.AOI_Name_Left)
    
    df['AOI_Name_Right'] = df['AOI Name Right']
    df['AOI_Name_Right'] = pd.Categorical(df.AOI_Name_Right)
    
    df['Trial'] = pd.Categorical(df.Trial)
    df['Stimulus'] = pd.Categorical(df.Stimulus)
    df['Color'] = pd.Categorical(df.Color)
    df['Class'] = pd.Categorical(df.Class)
    df['Gender'] = pd.Categorical(df.Gender)

    
    df = df.drop(df.loc[df['Category Group']==0].index, inplace=True)
    return df

proc_data(df1)
```
